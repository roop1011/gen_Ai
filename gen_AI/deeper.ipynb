```python
!pip install fbprophet
from fbprophet import Prophet
import pandas as pd

# Load demand data
df = pd.read_csv('demand_data.csv')  # Columns: 'date', 'demand'

# Prepare the data
df.columns = ['ds', 'y']  # Prophet expects 'ds' as the date and 'y' as the target variable

# Initialize the model
model = Prophet()

# Fit the model
model.fit(df)

# Make future predictions (365 days into the future)
future = model.make_future_dataframe(periods=365)
forecast = model.predict(future)

# Plot the forecast
model.plot(forecast)
```
